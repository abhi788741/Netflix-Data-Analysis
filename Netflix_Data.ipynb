```python
movies_df = df[df['type'] == 'Movie'].copy()
movies_df = movies_df.dropna(subset=['duration'])
movies_df = movies_df[movies_df['duration'].str.contains('min', na=False)]
movies_df['duration_minutes'] = movies_df['duration'].str.extract(r'(\d+)')
movies_df = movies_df.dropna(subset=['duration_minutes'])
movies_df['duration_minutes'] = movies_df['duration_minutes'].astype(int)
average_duration = movies_df['duration_minutes'].mean()
print("Average movie duration:",round(average_duration,2),"minutes")
```

```python
average_duration=99.58
avg_series=pd.Series([average_duration],index=['Average movie duration'])
avg_series.plot(kind='bar',figsize=(6,4))
plt.title("Average duration of movies")
plt.show()
```

```python
movies_df['duration_minutes'].plot(kind='hist',bins=20,color='green')
plt.title("distribution of Movie durations")
plt.show()
```
